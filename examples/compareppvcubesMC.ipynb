```python
# This is an example of the comparison of data cubes using the HOG technique
#
# Prepared by Juan D. Soler (soler@mpia.de)

import sys
sys.path.append('../')
from astrohog2d import *
from astrohog2d1v import *

import matplotlib.pyplot as plt
from astropy.io import fits

from scipy import ndimage
```

```python
# Load the cubes that you plan to compare
# Just for reference, I assume that the first index runs over the non-spatial coordinate
# Also assume that the cubes are spatially aligned and are reprojected into the same grid 

hdul = fits.open('../data/testcube1.fits')
cube1=hdul[0].data[:,0:512,512-256:512+256] # Cutting cube to make it run faster
hdul.close()
hdul = fits.open('../data/testcube2.fits')
cube2=hdul[0].data[:,0:512,512-256:512+256] # Cutting cube to make it run faster
hdul.close()
```

```python
fig, ax = plt.subplots(1,2, figsize=(12., 8.))
ax[0].imshow(cube1.sum(axis=0), origin='lower', cmap='terrain')
ax[0].set_title('Integrated cube 1')
ax[1].imshow(cube2.sum(axis=0), origin='lower', cmap='ocean')
ax[1].set_title('Integrated cube 2')
plt.show()
```

```python
# Here you select the size of your derivative kernel in pixels
ksz=10
```

```python
# Here I define the masks for both cubes
# For the sake of simplicity, I'm only masking the edges of the cubes
sz1=np.shape(cube1)
sz2=np.shape(cube2)
#mask=np.zeros(cube1)
#mask[:,0:ksz,:]=0.
#mask[:,sz1[1]-1-ksz:sz1[1]-1,:]=0.
#mask[:,:,0:ksz]=0.
#mask[:,:,sz1[2]-1-ksz:sz1[2]-1]=0.
```

```python
# Here you define the channel ranges over which you want to compare the cubes
zmin1=0
zmax1=sz1[0]-1
zmin2=0
zmax2=sz2[0]-1
```

```python
s_cube1=np.std(cube1)*np.ones_like(cube1)
s_cube2=np.std(cube2)*np.ones_like(cube2)
```

```python
# Run the HOG

#corrplane12, corrcube12, scube1, scube2 = HOGcorr_ppvcubes(cube1, cube2, zmin1, zmax1, zmin2, zmax2, ksz=ksz, mask1=mask, mask2=mask, s_cube1=s_cube1, s_cube2=s_cube2)
corrplane12, corrcube12, scube1, scube2 = HOGcorr_ppvcubes(cube1, cube2, zmin1, zmax1, zmin2, zmax2, ksz=ksz)                                                           
                                                           
```

```python
vplane=corrplane12['V']
s_vplane=corrplane12['s_V']
vdplane=corrplane12['Vd']
s_vdplane=corrplane12['s_Vd']
```

```python
fig, ax = plt.subplots(1,2, figsize=(14., 8.))
im0=ax[0].imshow(vplane, origin='lower', cmap='inferno')
ax[0].set_title(r'$V$')
plt.colorbar(im0, ax=ax[0], fraction=0.046, pad=0.04)
ax[1].imshow(vdplane, origin='lower', cmap='inferno')
ax[1].set_title(r'$V_{d}$')
plt.colorbar(im0, ax=ax[1], fraction=0.046, pad=0.04)
plt.show()
```
